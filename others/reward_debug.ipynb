```python
import sys
import traceback
sys.path.append('/mnt/cpfs2/rzhang/tools/dev_patch/project_c_grpo')

from src.utils.tools import get_pred_judgement_result


def math_format_reward(completions, **kwargs):

    rewards = []
    for completion in completions:
        try:
            _completion = completion[0]["content"]
            _has_answer, _ = get_pred_judgement_result(_completion)
            if _has_answer: 
                rewards.append(1.0)
            else:
                rewards.append(-1.0)
        except Exception as e:
            print(traceback.format_exc())
            rewards.append(-1.0)

    return rewards

math_format_reward(data["completions"])
```
